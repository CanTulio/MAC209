```python
#Bibliotecas utilizadas
import pandas as pd
import matplotlib.pyplot as plt
import glob
import numpy as np
import math
```

Abaixo estão as funções que foram utilizadas para plotar os dados simulados pelo método de euler

```python
# Funções utilizadas para Euler

def initState(x0, v0, t0):
    ## Funçao que retorna um estado inicial contendo a posição,
    # velocidade e tempo inicial, contidos num array numpy
    
    S = np.array([x0, v0, t0])
    return S

def rate(S, a, dt):
    ## Retorna um np.array contendo a taxa de variação do espaço
    #(v, ou seja, s[1]), velocidade (aceleração) 
    # e tempo (que é constante e vale 1)
    
    R = np.array([S[1], a, 1])
    return R

def nextState(S, a, dt):
    # Recebe um estado S e retorna o proximo estado Sn
    Sn = S + rate(S, a, dt)*dt
    return Sn
    
```

## Simulação

Para simular o movimento, analisamos a dinâmica das forças num pêndulo. Separamos em tópicos como obtivemos a atualização das seguintes grandezas físicas

### Velocidade angular


Fazendo uso de algumas equações do movimento circular, obtemos : 

$$ ac = \omega L \therefore \omega = \frac{a_c}{L} = \frac{-9.8sen(\theta)}{L} $$

### Aceleração centrípeta

Para determinar como atualizar a aceleração centrípeta, fizemos uso de uma análise da dinâmica do pêndulo. Pela segunda lei de newton, vale que

$$sin(\theta) = \frac{a_c}{g} \implies a_c = g * sin(\theta)$$ Por conveniência de orientação do sistema, multiplicamos o lado direito da equação obtendo 

$$a_c = -gsin(\theta) = -9.8sin(\theta)$$

### Ângulo

Para atualizar o ângulo, simplesmente somamos um valor de ângulo a uma pequena variação da velocidade angular



```python
def nextS(s,r,dt):
    #s = s + r*dt
    r[3] = -9.8*math.sin(s[2])/L
    
    r[2] = r[2] + r[3]*dt
    s[2] = s[2] + r[2]*dt
    
    s[0] = L*math.sin(s[2])
    s[1] = -L*math.cos(s[2])
    s[4] = s[4] +r[4]*dt
    
def eulerPendulo(x0,y0,teta,hx,hy,ha,ht,rw,ra):
    # s= [ x, y,             teta,                0, t]
    # r= [vx,vy,velocidadeAngular,aceleraçãoAngular, 1]
    s = [x0,y0,teta,0,0]
    r = [0,0,0,0,1]
    dt = 0.01
    
    for i in range(0,500):
        nextS(s,r,dt)
        hx.append(s[0])
        hy.append(s[1])
        ha.append(s[2])
        ht.append(s[4])
        rw.append(r[2])
        ra.append(r[3])

hx = []
hy = []
ha = []
ht = []
rw = []
ra = []

L = 0.8 # Comprimento da haste do pêndulo
teta = math.radians(-10) #Angulo inicial do movimento

#Simulação
eulerPendulo(L*math.sin(teta), -L*math.cos(teta),teta,hx,hy,ha,ht,rw,ra)

#Sx x t
fig = plt.figure(figsize = (50,4))
plt.subplot(131)
plt.plot(ht,hx,color = 'C0')
plt.xlabel("Tempo(s)")
plt.ylabel("Posição (rad)")
plt.grid(True)
plt.title("Posição no Eixo X pelo tempo ")

#Sy x t
fig = plt.figure(figsize = (50,4))
plt.subplot(131)
plt.plot(ht,hy,color = 'C1')
plt.xlabel("Tempo(s)")
plt.ylabel("Posição no eixo y (rad)")
plt.grid(True)
plt.title("Posição no Eixo Y pelo tempo ")

#w,a x t
fig = plt.figure(figsize = (50,4))
plt.subplot(131)
plt.plot(ht,rw,color = 'C2',label = "Velocidade Angular (rad/s)")
plt.plot(ht,ra,color = 'C3',label = "Aceleração Angular (rad/s^2)")
plt.xlabel("Tempo(s)")
plt.title("Velocidade e Aceleração Angulares pelo tempo ")
plt.legend()
plt.grid(True)
```

## Simulação

Tendo em vista a desaceleração do movimento, podemos modelar o experimento de forma que a aceleração seja determinada pela equação diferencial : 

$$
\frac{dv}{dt} = \alpha*v(t)
$$
Uma solução particular para tal equação seria 
$$
v(t) = e^{\alpha\cdot-t}
$$

Ou seja, a aceleração pode ser determinada como 

$$
a(t) = -\alpha e ^{-\alpha t}
$$

Sendo $\alpha = a(0)$

Ademais, pelos dados médios do experimento, vamos estimar que $a(0) = \alpha = 0.2$ e podemos estimar a velocidade angular inicial - em média - como 2 rad/s.

Com esses dados, podemos analisar como a velocidade se comporta num ambiente com atrito que é proporcional a velocidade do corpo, ou seja, cujo atrito diminui conforme o corpo desacelera.

```python
def acc(t, alpha):
    #Função que atualiza a aceleração
    return -alpha*math.exp(-alpha*t)

t0 = 0 
dt = 0.1
tVal = []
tf = 16 #Tempo Final de 16 segundos

v0 = 2 #velocidade inicial estimada
vVal = []

theta0 = 0
thetaVal = []

alpha = 0.2 #Valor estimado de alfa
a = -math.exp(0)
acVal = []

S = initState(theta0, v0, t0)

while(S[2] < tf):
    S = nextState(S, a, dt)
    a = acc(S[2],alpha)
    acVal.append(a)
    thetaVal.append(S[0])
    vVal.append(S[1])
    tVal.append(S[2])
    
#Gráfico Dist x t
fig = plt.figure(figsize=(50,4))
plt.subplot(131)
plt.plot(tVal, thetaVal)
plt.xlabel("Tempo(s)")
plt.ylabel("teta(t) (rad)")
plt.title("Distância Percorrida Pelo tempo")
plt.grid(True)
plt.show()


# W x t
fig = plt.figure(figsize=(50,4))
plt.subplot(131)
plt.plot(tVal, vVal)
plt.xlabel("Tempo(s)")
plt.ylabel("Velocidade angular (rad/s)")
plt.title("Velocidade Angular do Celular")
plt.grid(True)
plt.show()

#AcAng x t
fig = plt.figure(figsize=(50,4))
plt.subplot(131)
plt.plot(tVal, acVal, color = 'C3')
plt.xlabel("Tempo(s)")
plt.title("Atrito Incidente no Movimento")
plt.ylabel("Atrito (rad/s^2)")
plt.grid(True)
plt.show()
```

### Análise da simulação

Pelos gráficos, pode-se concluir que modelar a taxa de variação da velocidade proporcional à propria velocidade funciona bem, tendo em vista que os dados experimentais e teoricos se assemelham muito no intervalo $t\in [6, 15]$, além disso, presumir que o atrito tende a diminuir conforme o carro gira mais lentamente também faz bastante sentido, de forma que o resultado observado no gráfico de atrito é fiel à realidade. Pelo gráfico de angulação, nota-se que o banco girou ao menos 5 vezes e que sua velocidade de rotação não mudou muito durante o período : de fato, pelos dados simulados, a velocidade cai de 2rad/s para 1 rad/s em 15 segundos, ou seja, uma mudança sutil.

## Simulação

Utilizamos um Modelo de variação da aceleração extremamente parecido com o utilizado no movimento circular, com a adaptação das unidades para um movimento retilíneo. O valor escolhido para $\alpha$ foi de 0.0005, e ele também foi estimado com base nos dados.

```python
#Setup

t0 = 0
v0 = 0
s0 = 0
dt = 0.01
theta = math.radians(4)

g = 9.7      # g no laboratório didatico do IFUSP, medido pelo IAG, em m/s^2
at = g*math.sin(theta) # aceleração teorica no plano inclinado
```

```python
#SEM  ATRITO

tVals = []
sVals = []
vVals = []

S = initState(s0, v0, t0)

while(S[0] < 1.2): #Fim da rampa
    S = nextState(S, at, dt)
    sVals.append(S[0])
    vVals.append(S[1])
    tVals.append(S[2])
```

```python
#Com Atrito

tValsAt = []
sValsAt = []
vValsAt = []
aVals = []

px = g*math.sin(theta)

alpha = 0.001
a = px
S = []
S = initState(s0, v0, t0)

while(S[0] < 1.2): #Fim da rampa
    S = nextState(S, a, dt)
    a = a + acc(S[2],alpha)
    sValsAt.append(S[0])
    vValsAt.append(S[1])
    tValsAt.append(S[2])
    aVals.append(a)
```

```python
#Comparação

#S x t
fig = plt.figure(figsize=(50,4))
plt.subplot(131)
plt.grid(True)
plt.plot(tVals,sVals,label = "Sem Atrito")
plt.plot(tValsAt,sValsAt, label ="Com Atrito")
plt.title("Posição do Celular")
plt.ylabel("Espaço(m)")
plt.xlabel("Tempo (s)")
plt.legend()
plt.show()

#V x t
fig = plt.figure(figsize=(50,4))
plt.subplot(131)
plt.grid(True)
plt.plot(tVals,vVals, label ="Sem Atrito")
plt.plot(tValsAt,vValsAt, label ="Com Atrito")
plt.title("Velocidade do Celular")
plt.ylabel("Velocidade (m/s)")
plt.xlabel("Tempo (s)")
plt.legend()
plt.show()

#A x t
fig = plt.figure(figsize=(50,4))
plt.subplot(131)
plt.grid(True)
plt.plot(tValsAt,aVals, color = 'C3')
plt.title("Aceleração do Celular")
plt.ylabel("Aceleração (m/s^2)")
plt.xlabel("Tempo (s)")
plt.show()
```

A comparação revela justamente o comportamento esperado dos dois movimentos, realçando o fato do atrito ser muito pequeno. O atrito pequeno foi escolhido como candidato para a modelagem devido as condições de realização do movimento.

No mais o comportamento Parabólico da posição conforme a velocidade aumenta pode ser observado nas duas simulações.
